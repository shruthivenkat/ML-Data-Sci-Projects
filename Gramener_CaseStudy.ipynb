```python
# Taking only data which has loan status as charged off
loan_ChargeOff = loan[loan.loan_status == 'Charged Off']
```

```python
#Found mixed types of columns like term,int_rate,emp_length,rev_util
loan_ChargeOff.term = loan_ChargeOff.term.apply(lambda x: 60 if x  == ' 60 months' else 36 )

loan_ChargeOff.int_rate = loan_ChargeOff.int_rate.apply(lambda x: x.split('%')[0]) 

loan_ChargeOff.emp_length = loan_ChargeOff.emp_length.astype(str).apply(lambda x: x.split(' ')[1] if len(x.split(' ')) == 3 else x.split(' ')[0])
loan_ChargeOff.emp_length = loan_ChargeOff.emp_length.astype(str).apply(lambda x: x.split('+')[0])

loan_ChargeOff.revol_util = loan_ChargeOff.revol_util.astype(str).apply(lambda x: x.split('%')[0]) 
```

```python
loan_ChargeOff.info()
```

```python
#date conversion 
loan_ChargeOff.issue_d = pd.to_datetime(loan_ChargeOff.issue_d, format='%b-%y')

loan_ChargeOff.earliest_cr_line = pd.to_datetime(loan_ChargeOff.earliest_cr_line, format='%b-%y')

loan_ChargeOff.last_pymnt_d = pd.to_datetime(loan_ChargeOff.last_pymnt_d, format='%b-%y')

loan_ChargeOff.last_credit_pull_d = pd.to_datetime(loan_ChargeOff.last_credit_pull_d, format='%b-%y')
```
